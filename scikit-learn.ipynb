# 前処理

```python
import numpy as np
import pandas as pd
```

```python
# サンプルのデータセット
df = pd.DataFrame(
  {
    'A': [1, np.nan, 3, 4, 5],
    'B': [6, 7, 8, np.nan, 10],
    'C': [11, 12, 13, 14, 15]
  }
)
df
```

```python
df.isnull()
```

```python
# 欠損値の補完
from sklearn.impute import SimpleImputer
# 平均値で欠損値を補完するインスタンスを作成する
imp = SimpleImputer(strategy='mean')
# 欠損値を補完
imp.fit(df)
imp.transform(df)
```

```python
import pandas as pd
df = pd.DataFrame(
  {
    'A': [1, 2, 3, 4, 5],
    'B': ['a', 'b', 'a', 'b', 'c']
  }
)
df
```

## カテゴリ変数のエンコーディング

```python
from sklearn.preprocessing import LabelEncoder
# ラベルエンコーダーのインスタンスを生成
le = LabelEncoder()
le.fit(df['B'])
le.transform(df['B'])
```

```python
# 元の値を確認
le.classes_
```

## One-hotエンコーディング

```python
df
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

# B列（1番目のインデックス）にのみOneHotEncoderを適用
# remainder='passthrough' で他の列を残す
ct = ColumnTransformer(
    transformers=[("onehot", OneHotEncoder(), ['B'])],
    remainder='passthrough'
)

# 直接変換
df_ohe_array = ct.fit_transform(df)
df_ohe_array
```

# 特徴量の正規化

```python
import pandas as pd

df = pd.DataFrame(
  {
    'A': [1, 2, 3, 4, 5],
    'B': [100, 200, 400, 500, 800]
  }
)

df
```

```python
df['A'].mean()
```

```python
df['A'].std(ddof=0)
```

```python
# 分散正規化
from sklearn.preprocessing import StandardScaler
# 分散正規化のインスタンスを生成
stdsc = StandardScaler()
# 分散正規化を実行
stdsc.fit(df)
stdsc.transform(df)
```

```python
# 最小最大正規化
from sklearn.preprocessing import MinMaxScaler
# 最小最大正規化のインスタンスを生成
mmsc = MinMaxScaler()
# 最小最大正規化を実行
mmsc.fit(df)
mmsc.transform(df)
```
