```python
import matplotlib.style

# ggplotスタイルを指定
matplotlib.style.use('ggplot')
```

## MATLABスタイル

```python
x = [1, 2, 3]
y = [2, 4, 9]

plt.plot(x, y) # 折れ線グラフを描画
plt.title('MATLAB-style')

plt.show()
```

## オブジェクト指向スタイル

```python
x = [1, 2, 3]
y = [2, 4, 9]

# 描画オブジェクト(fig)とサブプロット(ax)を生成
fig, ax = plt.subplots()

ax.plot(x, y) # 折れ線グラフを描画
ax.set_title('OOP-style')

plt.show()
```

# 描画オブジェクト

## 描画オブジェクトとサブプロット

```python
fig, axes = plt.subplots(2)
plt.show()
```

```python
fig, axes = plt.subplots(2, 2) # 2行2列のサブプロットを配置
plt.show()
```

```python
fig, axes = plt.subplots(ncols=2) # 1行2列のサブプロットを配置
plt.show()
```
