## 変形(reshape)

```python
c1 = np.array([0, 1, 2, 3, 4, 5])
c1
```

```python
c2 = c1.reshape((2, 3))
c2
```

```python
# 元に戻す
c3 = c2.ravel() # 参照を返す
c3
```

```python
c4 = c2.flatten() # copyを返す
c4
```

```python
# ravelメソッドとflattenメソッド
c2
```

```python
c3 = c2.ravel() # 参照（浅いコピー）
c4 = c2.flatten() # コピー（深いコピー）
c3[0] = 6
c4[1] = 7
```

```python
c3
```

```python
c4
```

```python
c2
```

```python
# Python標準のリストは、スライスした結果はコピー（深いコピー）
p_list1 = [0, 1]
p_list2 = p_list1[:]
p_list2[0] = 2

print(p_list1)
print(p_list2)
```

```python
# NumPyのndarrayでのスライスした結果は参照
np_array1 = np.array([0, 1])
np_array2 = np_array1[:]
np_array2[0] = 2

print(np_array1)
print(np_array2)
```

## 乱数

```python
f = np.random.random((3, 2))
f
```

```python
np.random.seed(123)
np.random.random((3, 2))
```

```python
np.random.seed(123)
np.random.rand(4,2)
```

```python
np.random.seed(123)
np.random.randint(1, 10)
```

```python
np.random.seed(123)
np.random.randint(1, 10, (3, 3))
```

```python
np.random.seed(123)
# 0.0以上、5.0未満の2x3の行列の2次元配列
np.random.uniform(0.0, 5.0, size=(2, 3))
```

```python
np.random.seed(123)
np.random.uniform(size=(4, 3))
```

```python
np.random.seed(123)
# 標準正規分布から4x2の行列の2次元配列を得る
np.random.randn(4, 2)
```

```python
np.random.seed(123)
np.random.normal(50, 10, size=(3, 3))
```
