## ユニバーサルファンクション

```python
# 通常のPythonでの実装
li = [[-3, -2, -1],
      [0, 1, 2]]
new = []
for i, j in enumerate(li):
  new.append([])
  for k in j:
    new[i].append(abs(k))
```
