## 変形(reshape)

```python
c1 = np.array([0, 1, 2, 3, 4, 5])
c1
```

```python
c2 = c1.reshape((2, 3))
c2
```

```python
# 元に戻す
c3 = c2.ravel() # 参照を返す
c3
```

```python
c4 = c2.flatten() # copyを返す
c4
```
